```python
from pylab import *

alist = linspace(-1.5, 1.5, 200)
blist = linspace(-1.5, 1.5, 200)
u, w = meshgrid(alist, blist)
V = log((w+1+sqrt(u**2 + w**2+2*w+1))/(w-1+sqrt(u**2 + w**2-2*w+1)))

figure()
CP1 = contour(u, v, V)
clabel(CP1, fontsize=10) 
title('Contour Plot')
xlabel('u')
ylabel('w')
show()
```

```python
from pylab import *

alist = linspace(-1.5, 1.5, 200)
blist = linspace(-1.5, 1.5, 200)
u, v = meshgrid(alist, blist)
V = log((1+sqrt(u**2 + v**2+1))/(-1+sqrt(u**2 + v**2+1)))

figure()
CP1 = contour(u, v, V)
clabel(CP1, fontsize=10) 
title('Contour Plot')
xlabel('u')
ylabel('v')
show()
```
